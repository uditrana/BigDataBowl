## 2. Model Training

### 2 a. Splitting the data into test and train

```python


a['label'] = a['yac']
# lol = a['yac'].value_counts().plot.bar()
a.loc[a.label < -5, 'label'] = -5
#a.loc[a.label > 30, 'label'] = 30


a.loc[((a.label >= -5) & (a.label <= 0)), 'label'] =  0 #-2.5
a.loc[((a.label > 0) & (a.label <= 5)), 'label'] = 1 #2.5
a.loc[((a.label > 5) & (a.label <= 10)), 'label'] = 2 #7.5
a.loc[((a.label > 10) & (a.label <= 15)), 'label'] = 3 #12.5
a.loc[((a.label > 15) & (a.label <= 20)), 'label'] = 4 #17.5
a.loc[((a.label > 20) & (a.label <= 25)), 'label'] = 5 # 22.5
a.loc[((a.label > 25) & (a.label <= 30)), 'label'] = 6 # 27.5
a.loc[(a.label > 30), 'label'] = 7 #30.5

#a["label"] = a['label'] + 2.5
print(a['label'].unique())
# First we drop the columns we dont want. 
b_to_train = a.drop(columns=['gameId', 'playId', "week", "yac", "label", "xyac_nflfastR"]) # these are not columns we want to use to train

# Now we split based on number, first 3 weeks have 2250 plays, we will use 2000

b_yac = a["label"]



train = b_to_train[:4000]
train_label = b_yac[:4000]

test = b_to_train[4000:]
test_label = b_yac[4000:]

big_test = test.append([test]*40, ignore_index=True)
big_label = test_label.append([test_label]*40, ignore_index=True)
        

dtrain = xgb.DMatrix(train, label = train_label)
dtest = xgb.DMatrix(test, label = test_label)

dtestBig = xgb.DMatrix(big_test, label = big_label)

print(len(big_test))
```

### 2 b. Training the Model on the train set:

```python
param = {'max_depth': 4, 'eta': 0.3, 'objective': 'multi:softprob', "num_class" : 8,
         "eval_metric" : "mlogloss", "booster" : "gbtree", "gamma" : .5, "subsample" : 0.8,
        "colsample_bytree" : 0.8, "min_child_weight" : 1, "lambda" : 1.3}
bst = xgb.train(param, dtrain, 400, verbose_eval = 1)
```

```python
model_data = pd.read_csv("/Users/Rishav/Documents/2021bdbwinners/epa_model_data.csv")
model_labels = model_data['label']
model_weights = model_data['Total_W_Scaled']
model_data_correct_calls = model_data.drop(columns=['label', 'Total_W_Scaled', 'Unnamed: 0', "season", "half_seconds_remaining"])

print(len(model_data_correct_calls.columns))
epa_train = xgb.DMatrix(model_data_correct_calls, label = model_labels, weight = model_weights)
```
